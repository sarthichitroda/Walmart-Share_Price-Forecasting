```python
#Importing all the necessary libraries to ensure smooth working of the codes

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from fbprophet import Prophet
```

```python
#Importing Dataset

mydata=pd.read_csv('WMT.csv')
```

```python
mydata.head() #Displaying initial 5 rows of the dataset
```

```python
#Plotting the values of Close column using Matplot library

close=mydata['Close']

y=close.plot(title='Walmart')
y.set_xlabel('Date')
y.set_ylabel('Closing Price')
plt.show()
```

```python
mydata['Date']=pd.to_datetime(mydata["Date"], infer_datetime_format=True)
mydate=mydata[["Date", "Close"]]
```

```python
mydata=mydata.rename(columns={'Date':'ds', "Close":'y'})
```

```python
forecastmodel=Prophet()
forecastmodel.fit(mydata)
wmtforecast=forecastmodel.make_future_dataframe(periods=365)
predict=forecastmodel.predict(wmtforecast)
predict[["ds","yhat","yhat_lower","yhat_upper"]].tail()
```

```python
graph=forecastmodel.plot(predict, xlabel="Date", ylabel="Price")
```

```python
graph1=forecastmodel.plot_components(predict)
```
